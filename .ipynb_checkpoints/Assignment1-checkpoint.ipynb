# Part 1: Binning degree distributions


**Binning real numbers**
* Let's do a gentle start and use the `random` library to generate 5000 data points from a Gaussian distribution with $\mu = 2$ and $\sigma = 0.125$.

```python
mu, sigma, N = 2, 0.125, 5000
points = [rd.gauss(mu,sigma) for _ in range (N)]
```

  * Now, let's use `numpy.histogram` to bin those number into 10 bins.

```python
histogram = np.histogram(points, 10)
histogram
```

```python
print(f"Length of histogram's arrays: ({len(histogram[0])},{len(histogram[1])})")
```

**Question: What does the `numpy.histogram` function return? Do the two arrays have the same length?**

*The `numpy.histogram` returns a tuple of 2 arrays: the number of values in each bin, then bins' intervals. The two arrays don't have the same length. Indeed, the second array is one value longer since each interval is described by 2 bounds.*

 * Then we use `matplotlib.pyplot.plot` to plot the binned data.

```python
x = [ (histogram[1][i+1] + histogram[1][i])/2 for i in range (len(histogram[1])-1)]
plt.plot(x,histogram[0])
plt.title("Binned data plot")
```

**Question: You will have to deal with the fact that the counts- and bin-arrays have different lengths. Explain how you deal with this problem and why.**

*To deal with the different length problem, we can replace intervals by the center point of each interval for x-axis. Thus, we obtain 10 center points for 10 y-axis values*

**Binning integers**
  * But binning real numbers into a fixed number of bins is easy when `numpy.histogram` does all the work and finds the right bin boundaries for you. 
Now we'll generate a bunch of integers and set the bin boundaries manually. This time, let's grab data from a Poisson distribution. As it turns out `numpy` also has some convenient random number generators. Use `numpy.random.poisson` to generate 5000 numbers drawn from a Poisson distribution characterized by $\lambda = 10$. Find the maximum and minimum value of your 5000 random numbers. 

```python
# Generating 500 numbers drawn from Poisson distribution
lambdaa, N = 10, 5000
points = [np.random.poisson(lambdaa) for _ in range (N)]
```

```python
# Find max and min values
points_max, points_min = max(points),min(points)
print(f"(max value, min value) = ({points_max}, {points_min})")
```

  * Instead of simplify specifying the number of bins for `numpy.histogram`, let's specify the bins we want using a vector. Create a vector $v$ that results in a binning that puts each integer value in its own bin and where the first bin contains the minimum number you found above, and the last bin contains the maximum number. 

```python
v = list(range(points_min,points_max+1))
```

**Question:  you may want to think about how `numpy.histogram` deals with the smallest an largest value in particular; is it correct to set the first entry of $v$ to be equal to the smallest value you observe and the last entry of $v$ to be the highest value observed?**

*Yes, it is correct to process this way since all values are found between those two bounds.*

  * Use the vector by setting `numpy.histogram`'s `bin` parameter as `bin = ` $v$.

```python
values, bins = np.histogram(points, bins=v)
```

**Binning and plotting degree distributions.**
  * Let's generate the Erd√∂s-Renyi (ER) network which has a degree distribution that matches the Poisson distribution above. 
First we have to figure out which values the ER parameters (_N_ and _p_) should assume. 

**Question: It's easy to see that $N = 5000$, but how do you find $p$?** (Hint: The parameter $\lambda$ in the Poisson distribution corresponds to the average degree, so you have to find a $p$ that results in an average degree, $k = 10$. And you know that $\langle k \rangle = p (N-1)$, which will give you $p$.)



  * Now, use `networkx` to create the graph and extract the degree distribution.

```python
N, k = 5000, 10
p = k/(N-1)

# ER graph creation
G_ER = nx.erdos_renyi_graph(N,p)
```

```python
# degree distribution extraction
all_degrees = [G_ER.degree(node) for node in list(G_ER.nodes)]
```

  * Finally, create a nice bar plot of the degree distribution, including axes labels and a plot title. Make sure that it looks like the Poisson distribution you plotted above.

```python
# Find max and min values
max_points, min_points = max(all_degrees), min(all_degrees)

# Construct v-vector as before
v = list(range(min_points, max_points + 1))

# Get histogram
values, bins = np.histogram(all_degrees, bins = v)

# Plot bar plot of the degree distribution and previous Poisson distribution for comparison
plt.figure(figsize=[20,4])

_ = plt.subplot(121)
plt.title(f"Erd√∂s-Renyi (ER) network's degree distribution, with N={N} and p={round(p,3)}")
plt.xlabel("Degree")
plt.ylabel("Number of concerned nodes")
plt.bar(x = bins[:-1], height = values)

_ = plt.subplot(122)
plt.bar(x=bins[:-1],height=values)
plt.title("Previously computed Poisson distribution")
```

*We note that both distributions are similar*

**Erd≈ës-R√©nyi Networks**

Consider an Erd≈ës-R√©nyi network with $N = 3000$ nodes, connected to each other with probability $p = 10^‚Äì3$.

**1. What is the expected number of links, ‚å©L‚å™?**


```python
p, N = 10**(-3), 3000
L_avg = p*N*(N-1)/2
print(f"‚å©L‚å™ = {L_avg}")
```

**2. In which regime is the network?**

Network **regimes**, based on probability $p$: 
- **subcritical**: $p$ < 1/$N$
- **supercritical**: $p$ > 1/$N$
- **connected**: $p$ > ln$N$/$N$

```python
print(f"1/N = {1/N}")
print(f"ln(N)/N = {np.log(N)/N}")
```

*Thus, based on network regime's definitions, the network is in **supercritical** regime.*

**3. Calculate the probability $p_c$ so that the network is at the critical point**


```python
pc = 1/N
print(f"pc = {pc}")
```

**4. Given the linking probability $p = 10^-3$, calculate the number of nodes $Ncr$ so that the network has only one component.**

We can read in the Network Science book that:
> Only the actor network is in the connected regime, meaning that all nodes are part of a single giant component. Note that while the boundary between the subcritical and the supercritical regime is always at ‚Äπk‚Ä∫ = 1, the boundary between the supercritical and the connected regime is at lnN, which varies from system to system. 

Thus, the network has only one component if ‚Äπk‚Ä∫ = ln(N). We know that ‚Äπk‚Ä∫ = p(N-1), so we have to **solve ln(N) = p(N-1)**.

Let's solve it graphically through approximation.

```python
x_points = list(range(1,10000))
y1 = [ np.log(x) for x in x_points]
y2 = [p*(x-1) for x in x_points]

plt.plot(x_points,y1)
plt.plot(x_points,y2)
plt.title("Curves to solve ln(N)=p.(N-1)")
plt.xlabel('N')
plt.legend(['ln(N)','p.(N-1)'])
```

```python
# Find the abscisse of intersection
i0=1
while i0<len(x_points):
    if (y2[i0]<y1[i0]):
        i0+=1
    else:
        break;
Ncr = x_points[i0]
print(f"Ncr = {Ncr}")
```

5. For the network in (d), calculate the average degree ‚å©kcr‚å™ and the average distance between two randomly chosen nodes ‚å©d‚å™.


```python
kcr_avg = p*(Ncr-1)
print(f"‚å©K_cr‚å™ = {kcr_avg}")
```

```python
d_avg = np.log(Ncr)/np.log(kcr_avg)
print(f"‚å©d‚å™ = {d_avg}")
```

```python
GCC = list(nx.connected_components(Grd))
GCC.sort(key=len)
concerned_nodes = GCC[-1]

GCC = Grd.subgraph(concerned_nodes).copy()

nx.draw(GCC,**options)
```

* Choose a node at random from the GCC. (Hint: You may want to try `random.choice`.)

```python
rd_node = rd.choice(list(GCC.nodes()))
print(f"We randomly chose the node n¬∞{rd_node}")
```

  * Find all nodes that are precisely 2 steps away from that node. (Hint. I suggest `networkx.single_source_shortest_path_length`)

```python
twosteps_nodes = list(set(nx.single_source_shortest_path_length(GCC,rd_node,2))-set(nx.single_source_shortest_path_length(GCC,rd_node,1)))
print(f"All nodes precisely 2 steps away from our node are : {list(twosteps_nodes)}")
```

  * Plot the GCC with the following choice of colors. Starting node _black_ (`"#000000"`). The nodes 2 steps away _red_ (`"#ff0000"`). All other nodes _blue_ (`"#A0CBE2"`). Again, I suggest using `networkx.draw()` and reading [the documentation](https://networkx.github.io/documentation/stable/reference/drawing.html) carefully find out how to color individual nodes.

```python
pos = nx.spring_layout(GCC)

options = {
    'node_size': 40,
    'edgecolors':'black'
}

fig, ax = plt.subplots(figsize=(8, 8))
plt.title("GCC of our random network with N=200 and <k>=1.5")
nx.draw_networkx_nodes(GCC,pos=pos, node_color='#A0CBE2', **options)
nx.draw_networkx_nodes(GCC,pos=pos, nodelist=twosteps_nodes, node_color='#ff0000', **options)
nx.draw_networkx_nodes(GCC,pos=pos, nodelist=[rd_node],node_color='black', **options)
nx.draw_networkx_edges(GCC,pos=pos)

plt.legend(['other nodes', '2 steps away nodes','chosen node'])
```

# Part 5 Power-laws and the friendship paradox

Next step is to explore the [Friendship paradox](https://en.wikipedia.org/wiki/Friendship_paradox). This paradox states that _almost everyone_ have fewer friends than their friends have, on average. This sounds crazy, but is actually an almost trivial consequence of living in a social network with a power-law degree distribution.
