**Erdős-Rényi Networks**

Consider an Erdős-Rényi network with $N = 3000$ nodes, connected to each other with probability $p = 10^–3$.

**1. What is the expected number of links, 〈L〉?**


```python
p, N = 10**(-3), 3000
L_avg = p*N*(N-1)/2
print(f"〈L〉 = {L_avg}")
```

**2. In which regime is the network?**

Network **regimes**, based on probability $p$: 
- **subcritical**: $p$ < 1/$N$
- **supercritical**: $p$ > 1/$N$
- **connected**: $p$ > ln$N$/$N$

```python
print(f"1/N = {1/N}")
print(f"ln(N)/N = {np.log(N)/N}")
```

*Thus, based on network regime's definitions, the network is in **supercritical** regime.*

**3. Calculate the probability $p_c$ so that the network is at the critical point**


```python
pc = 1/N
print(f"pc = {pc}")
```

**4. Given the linking probability $p = 10^-3$, calculate the number of nodes $Ncr$ so that the network has only one component.**

We can read in the Network Science book that:
> Only the actor network is in the connected regime, meaning that all nodes are part of a single giant component. Note that while the boundary between the subcritical and the supercritical regime is always at ‹k› = 1, the boundary between the supercritical and the connected regime is at lnN, which varies from system to system. 

Thus, the network has only one component if ‹k› = ln(N). We know that ‹k› = p(N-1), so we have to **solve ln(N) = p(N-1)**.

Let's solve it graphically through approximation.

```python
x_points = list(range(1,10000))
y1 = [ np.log(x) for x in x_points]
y2 = [p*(x-1) for x in x_points]

plt.plot(x_points,y1)
plt.plot(x_points,y2)
plt.title("Curves to solve ln(N)=p(N-1)")
plt.xlabel('N')
_ = plt.legend(['ln(N)','p(N-1)'])
```

```python
# Find the abscisse of intersection
i0=1
while i0<len(x_points):
    if (y2[i0]<y1[i0]):
        i0+=1
    else:
        break;
Ncr = x_points[i0]
print(f"Ncr = {Ncr}")
```

**5. For the network in (d), calculate the average degree 〈kcr〉 and the average distance between two randomly chosen nodes 〈d〉.**


```python
kcr_avg = p*(Ncr-1)
print(f"〈K_cr〉 = {kcr_avg}")
```

```python
d_avg = np.log(Ncr)/np.log(kcr_avg)
print(f"〈d〉 = {d_avg}")
```

# Part 3: Watts-Strogatz Networks



** Use nx.watts_strogatz_graph to generate 3 graphs with 500 nodes each, average degree = 4, and rewiring probablity  p= 0, 0.1,and 1**.

```python
# Importing all the necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import random as rn
import networkx as nx


N = 500  # nodes
k =  4  # average degree
p = [0, 0.1, 1]  # rewiring probability

# Generating three Watts-Strogatz graphs with same number of nodes 
# and same average degree but different rewiring probability

WS = [] # list of graphs
for j in range(len(p)):
    WS.append(nx.watts_strogatz_graph(N, k, p[j]))
    
```

** Calculate the average shortest path length  ⟨d⟩  for each one. **

```python
d = []  # list of <d>
for i in range(len(WS)):    
    d.append(nx.average_shortest_path_length(WS[i]))

print('The average shortest path length <d> for each Watts-Strogatz graph is:')
print(d)
```

** Describe what happens to the network when  p=1. **

For p=1, all links are rewired and the model turns into a fully randomized network. The network has the shortest average length path, representing the small-world phenomena. However the clustering is getting low.

Let's understand the behavior of the WS model as we increase p in more detail. Generate 50 networks with N=500, ⟨k⟩=4, for each of p = {0,0.01,0.03,0.05,0.1,0.2}.

```python
N = 500  # nodes
k = 4  # average degree

WSL = [] # list of lists of networks for each p, len(WSL) = 6, len(WSL[0]) = 50
p = [0, 0.01, 0.03, 0.05, 0.1, 0.2]
for j in range(len(p)):
    WS = []  # list of 50 networks for specific p, len(WS) = 50
    for i in range(50):
        WS.append(nx.watts_strogatz_graph(N, k, p[j]))
    WSL.append(WS)
    
 
  
```


** Calculate the average of  ⟨d⟩ as well as the standard deviation over the 50 networks, to create a plot that shows how the path length decreases very quickly with only a little fraction of re-wiring. Use the standard deviation to add errorbars to the plot.  **   

```python
d_50 = np.array([]) # list of the average of <d> for each p 
std = np.array([]) # list of teh standard deviation for each p

for i in range(len(WSL)):
    d=np.array([])
    for j in WSL[i]:
        # the average shortest path length <d> for each network (for each p)
        d = np.append(d, nx.average_shortest_path_length(j)) 
        # len(d) = 50
    
    d_50 = np.append(d_50, np.mean(d))  
    std = np.append(std, np.std(d))
  
```

```python
d_50.tolist(), std.tolist()
```

```python
# Creating the plot 

plt.figure
plt.plot(p,d_50)
plt.title('Average shortest path length for 50 networks after rewiring')
plt.xlabel('Rewiring probability')
plt.ylabel('Average <d>')
plt.errorbar(p,d_50.tolist(), yerr = std.tolist(), xerr=None, capsize = 2, fmt = 'ro', label = 'standard deviation')
```

# Part 4. The Barabasi-Albert Model

We're going to create our own Barabasi-Albert model (a special case) in a `notebook`. Follow the recipe below for success

* Create a 100 node BA network using a BA model that you've coded on your own (so don't use the built-in NetworkX function, but the one you created during week 3). And plot it using NetworkX.


```python
# define options for following plots
options = {
    'node_size': 30,
    'edgecolors':'black',
    'node_color':'yellow'
}


def create_ba_graph(number_of_nodes):
    """the function generates a graph with the 
    desired number of nodes"""

    # First create a graph consisting of a single link
    G = nx.Graph()
    G.add_nodes_from([1,2])
    G.add_edge(1,2)

    #labels to be used for the graph
    new_node_label = 3

    # adding new nodes until desired number of nodes
    for _ in range (number_of_nodes-2):
        G.add_node(new_node_label)
        flatten_list = list(sum(G.edges,()))
        target_label = rd.choice(flatten_list)
        G.add_edge(new_node_label,target_label)
        new_node_label += 1

    return G
```

```python
# Creating a 100 node BA network
G1 = create_ba_graph(100)

# plot 100-node-graph
nx.draw_spring(G1, **options)
plt.show()
```

# Part 5 Power-laws and the friendship paradox

Next step is to explore the [Friendship paradox](https://en.wikipedia.org/wiki/Friendship_paradox). This paradox states that _almost everyone_ have fewer friends than their friends have, on average. This sounds crazy, but is actually an almost trivial consequence of living in a social network with a power-law degree distribution. The explanation is that almost everyone is friends with a hub, that drives up the average degree of the friends. Let's explore that in the 5000 node BA network we've just generated. Do the following:



* Pick a node _i_ at random (e.g. use `random.choice`). [Find its degree](http://networkx.lanl.gov/reference/generated/networkx.Graph.degree.html).


```python
# a random node i from the network 
i = rn.choice(list(G5k.nodes))

# i's degree
k_i = G5k.degree(i)

print('The random pick is node', i, 'having degree', k_i)
```

* Find _i_'s [neighbors](http://networkx.lanl.gov/reference/generated/networkx.Graph.neighbors.html). And calculate their average degree.

```python
# list of i's neighbors
i_nei = list(G5k.neighbors(i))

# i's neighbors degree
k_i_nei = np.mean(np.array(list((dict(G5k.degree(i_nei))).values())))

print('The neighbors are ', i_nei, 'having average degree', k_i_nei)
```

* Compare the two numbers to check if it's true that _i_'s friends (on average) have more friends than _i_.


```python
k_i < k_i_nei
```

* Do this 1000 times. How many out of those 1000 times is the friendship paradox true?


```python

def friendshipParadox(G):
    
    i=0 #counter
    
    for i in range(100):
        
        #Pick a node i at random
        pick = rn.choice(list(G.nodes))
        #Find its degree
        k_pick = G.degree(pick)
        
        
        #Find i's neighbors
        picks_nei = list(G.neighbors(pick))  ##or list(G[pick])
        #Calculate their average degree
        k_picks_nei = np.mean(np.array(list((dict(G.degree(picks_nei))).values())))
        
        #Compare them
        if (k_pick < k_picks_nei):
            i += 1 
        
    print("The friendship paradox is applied here " + str(i) + "%")
 

```

```python
friendshipParadox(G5k)
```
